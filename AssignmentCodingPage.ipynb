```bash
#fix the format

curl -s "https://restcountries.com/v3.1/name/{Cameroon}?&format=json" | jq
```

```bash
#Select the variables we're interested in and get a .csv: common name, official name, region, subregion, languages

curl -s "https://restcountries.com/v3.1/name/{Cameroon}" | jq '.[] | [.name.common, .name.official, .region, .subregion, (.languages | to_entries[] | .value)] | @csv'
```

```bash
#Now let's try it for multiple countries in an array

countries=("Cameroon" "Jordan" "Eritrea" "Congo" "Syria")

for X in "${countries[@]}"
do  
    curl -s "https://restcountries.com/v3.1/name/$X" | jq '.[] | [.name.common, .name.official, .region, .subregion, (.languages | to_entries[] | .value)] | @csv'
done
```
