```bash
while IFS= read -r line
do
    encoded_query=$(echo "$line" | jq -sRr @uri)
    response=$(curl -s "https://nominatim.openstreetmap.org/search?q=${encoded_query}&format=json")
    echo "$response" | jq '.[] | [.lat, .lon, .addresstype, .name]'
    sleep 1  # Delay to respect API rate limit
done < Kenya.csv
```

```bash
echo "Latitude,Longitude,AddressType,Name" > KenyaGeo_Test_2.csv

while IFS= read -r line
do
    encoded_query=$(echo "$line" | jq -sRr @uri)
    response=$(curl -s "https://nominatim.openstreetmap.org/search?q=${encoded_query}&format=json")
    echo "$response" | jq -r '.[] | [.lat, .lon, .addresstype, .name] | @csv' >> KenyaGeo_Test_2.csv
    sleep 1  # Delay to respect API rate limit
done < Kenya.csv
```

```bash
echo "Latitude,Longitude,AddressType,Name" > KenyaGeo_Test_3.csv

while IFS= read -r line
do
    encoded_query=$(echo "$line" | jq -sRr @uri)
    response=$(curl -s "https://nominatim.openstreetmap.org/search?q=${encoded_query}&format=json")
    echo "$response" | jq -r '.[] | [.lat, .lon, .addresstype, .name] | join(",")' >> KenyaGeo_Test_3.csv
    sleep 1  # Delay to respect API rate limit
done < Kenya.csv
```
